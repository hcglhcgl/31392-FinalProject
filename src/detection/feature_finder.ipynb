```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img_left = cv2.imread('../../rectified_images/left/200.png')
img_right = cv2.imread('../../rectified_images/left/200.png')

gray_left = cv2.cvtColor(img_left, cv2.COLOR_RGB2GRAY)
gray_right = cv2.cvtColor(img_right, cv2.COLOR_RGB2GRAY)
```

```python
# Initiate ORB detector
orb = cv2.ORB_create()
# find the keypoints with ORB
kp, des = orb.detectAndCompute(gray_left,None)
# draw only keypoints location,not size and orientation
img2 = cv2.drawKeypoints(gray_left, kp, None, color=(0,255,0), flags=0)
plt.imshow(img2)
```

```python
# find the keypoints with ORB
kp2, des2 = orb.detectAndCompute(gray_right,None)
# draw only keypoints location,not size and orientation
img3 = cv2.drawKeypoints(gray_right, kp2, None, color=(0,255,0), flags=0)
plt.imshow(img3)
```

```python
# Matching the keypoints
matcher = cv2.BFMatcher()
matches = matcher.match(des,des2)

matches = sorted(matches, key = lambda x:x.distance)

img3 = cv2.drawMatches(gray_left,kp,gray_right,kp2,matches[:10],None,flags=cv2.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS)

plt.figure(figsize = (20,20))
plt.imshow(img3)
```
